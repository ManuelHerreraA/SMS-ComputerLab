```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def polynomial_function(param, x):
    return param[0] * x**3 + param[1] * x**2 + param[2] * x + param[3]

param1 = [0.4, -3.0, 2.0, 10.0]
param2 = [-0.3, 1.0, 2.0, 2.0]
x = np.linspace(-2, 6, 100)
y1 = polynomial_function(param1, x)
y2 = polynomial_function(param2, x)


plt.figure(figsize=(10, 6))
plt.plot(x, y1, label=f'f(x) = {param1[0]}x + {param1[1]}x + {param1[2]}x + {param1[3]}')
plt.plot(x, y2, label=f'f(x) = {param2[0]}x + {param2[1]}x + {param2[2]}x + {param2[3]}')
plt.title('Polynomial Functions')
plt.xlabel('x')
plt.ylabel('y')


plt.grid(True, linestyle='--', alpha = 0.7) 
plt.axhline(0, color='gray', linewidth = 0.8)
plt.axvline(0, color='gray', linewidth = 0.8)
plt.legend()
#plt.savefig('PolynomialFunctions')
plt.show()
```

```python
# 1. Define Coeficient Matrix (A)
A = np.array([
    [1, 2, 3, 4],  # First equation coeficients
    [135, 88, 37, 13],  # Seconnd equation coeficients
    [-47, 22, -11, 8],  # Third equation coeficients
    [7, 15, -31, -43]   # Fourth equation coeficients
])

# 2. Define Right Hand Vector
b = np.array([22, 337, 332, 12])

# 3. Solve the Linear System
# np.linalg.solve(A, b) find the x Vector that satisfies Ax = b
try:
    solution = np.linalg.solve(A, b)

    print("\nSolution Vector (x):")
    print("x, y, z, t =", np.round(solution, 4))
    
    # Podemos asignar las soluciones a variables
    x1, x2, x3, x4 = solution

    # print(f"\nVerificacin:")
    # # Verificamos la primera ecuacin como ejemplo
    # verificacion1 = 1*x1 + 2*x2 + 3*x3 + 4*x4
    # print(f"1*x1 + 2*x2 + 3*x3 + 4*x4  {np.round(verificacion1, 4)} (Debe ser 22)")

except np.linalg.LinAlgError:
    print("\nThere is no unique solution).")
```

```python
# Define analytical function
def f(param, x):
    return param[0]*x**3 + param[1]*x**2 + param[2]*x + param[3]

def Df(param, x):
    return 3*param[0]*x**2 + 2*param[1]*x + param[2]

param1 = [0.4, -3.0, 2.0, 10.0]

# Central Derivative Method
def Central_Derivative(f, x, h = 1e-5):
    return (f(param1, x + h) - f(param1, x - h)) / (2*h)

x = np.linspace(-2, 6, 200)
y = f(param1, x)
Dy = Df(param1, x)

derivative_f_vectorized = np.vectorize(Central_Derivative)
y_derivative = derivative_f_vectorized(f, x)

plt.figure(figsize=(12, 7))

# Original function
plt.plot(x, y, label=f'f(x) = {param1[0]}x + {param1[1]}x + + {param1[2]}x + {param1[3]}', color='blue', linewidth=2)

# Analytical Derivative function
plt.plot(x, Dy, label=f'f\'(x) = {3*param1[0]:.1f}x^2 + {2*param1[1]}x + + {param1[2]} (Analytical Derivative)', color='green', linewidth=2, alpha = 0.7)

# Numerical Derivative function
plt.plot(x, y_derivative, label='f\'(x) (Numerical Derivative)', color='red', linestyle='--', linewidth=2)


plt.title('Polynomial Function and its Derivative')
plt.xlabel('x')
plt.ylabel('y')


plt.grid(True, linestyle='--', alpha = 0.7) 
plt.axhline(0, color='gray', linewidth = 0.8) 
plt.axvline(0, color='gray', linewidth = 0.8) 
plt.legend() 
plt.savefig('PolynomialFuncDeriv')
plt.show()
```
